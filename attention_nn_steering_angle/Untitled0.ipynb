```python
import torch
from torch import nn
```

```python
class Decoder(nn.Module):
  def __init__(self, alpha_t_size, y_size, h_size, embed_size):
    self.fout = nn.Linear(alpha_t_size, h_size)
    self.lstm = nn.LSTM(y_size, h_size, 1, batch_first=True)
    self.attention = nn.MultiheadAttention(embed_size , 6, batch_first=True)


  def forward(self, x):
    lstm_out, h_n, c_n = self.lstm(x)
    att_out, att_weights = self.attention(x, h_n) #lstm_out??
    pi = torch.flatten(torch.matmul(att_out, x))
    fsteer = self.fsteer(self.fout(pi))

    return att_out, fsteer


  def fsteer(self, u, v, d=1, k_s=1, k_slip=1):
    return u*d*k_s*(1+k_slip*(v**2))
```
